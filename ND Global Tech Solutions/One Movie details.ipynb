```python
import requests
from bs4 import BeautifulSoup as BS
sites="https://www.lyricstape.com/album/adipurush/1014"
sitei= requests.get(sites)
    
```

```python
sitesoup= BS(sitei.text,"html.parser")
print(sitesoup)
```

```python
tracks=sitesoup.find('div',class_="container one").find('div',class_="col-xs-12 col-sm-12 col-md-8 col-lg-8 col-xl-8")
print(tracks)
```

```python
albdet=sitesoup.find('div', class_="col-xs-12 col-sm-12 col-md-4 col-lg-4 col-xl-4")
print(albdet)
```

```python
mdlist=albdet.find('ul',class_="text-left")
print(mdlist)
```

```python
mname=albdet.find('ul',class_="text-left").find('a').text.strip()
print(mname)
```
